```python
import numpy as np
import pandas as pd
import matplotlib.image  as mpimg
import matplotlib.pyplot as plt

#total cases per day for globe and by country
df = pd.read_csv('total_cases (1).csv')
df.head()
```

```python
#total deaths by day for globe and country
df2 = pd.read_csv('total_deaths (1).csv')
df2.head()
```

```python
df2 = df2[['World', 'date']]
df2 = df2.dropna()
df2.head()
print(len(df2.index))
df2 = df2.rename(columns={'World': 'TotalDeaths'})
#deaths start 22 days in after cases start
df2 = df2.iloc[22:]
print(len(df2.index))
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
df = df[['World', 'date']]
df = df.dropna()
df = df.iloc[33:]

#save dataset without transformations
untouched = df
#merge in data about deaths
untouched = untouched.merge(df2)
untouched = untouched.set_index('date')

date = list(df['date'])
date2 = []
for i in range(len(date)):
    if (i+1) % 7 == 0:
        date2.append(date[i])
df = df.merge(df2)

#apply logarithmic transformation to data to remove multiplicative trend
df['World'] = np.log10(df["World"])
df['TotalDeaths'] = np.log10(df["TotalDeaths"])

df = df.set_index('date')
df.rolling(window = 7).mean().plot(figsize=(8,4), color="tab:red", title="Rolling Mean over 7 day period");
df.head()
```

```python
from statsmodels.tsa.stattools import adfuller

#show stats for total deaths before logarithmic transformation applied
#p-value is .99 which is much greater than .05, indicating data is not stationary
dftest = adfuller(untouched['TotalDeaths'], autolag = 'AIC')

print("1. ADF : ",dftest[0])
print("2. P-Value : ", dftest[1])
print("3. Num Of Lags : ", dftest[2])
print("4. Num Of Observations Used For ADF Regression and Critical Values Calculation :", dftest[3])
print("5. Critical Values :")
for key, val in dftest[4].items():
    print("\t",key, ": ", val)
```

```python
#show stats for total cases before logarithmic transformation applied
#p-value is .99 which is much greater than .05, indicating data is not stationary
dftest = adfuller(untouched['World'], autolag = 'AIC')

print("1. ADF : ",dftest[0])
print("2. P-Value : ", dftest[1])
print("3. Num Of Lags : ", dftest[2])
print("4. Num Of Observations Used For ADF Regression and Critical Values Calculation :", dftest[3])
print("5. Critical Values :")
for key, val in dftest[4].items():
    print("\t",key, ": ", val)
```

```python
#show stats for total cases after logarithmic transformation applied
#p-value is 1.55 x 10^-6 which is less than .05, indicating data is now stationary
dftest = adfuller(df['World'], autolag = 'AIC')

print("1. ADF : ",dftest[0])
print("2. P-Value : ", dftest[1])
print("3. Num Of Lags : ", dftest[2])
print("4. Num Of Observations Used For ADF Regression and Critical Values Calculation :", dftest[3])
print("5. Critical Values :")
for key, val in dftest[4].items():
    print("\t",key, ": ", val)
```

```python
#show stats for total deaths after logarithmic transformation applied
#p-value is .0009 which is less than .05, indicating data is now stationary
dftest = adfuller(df['TotalDeaths'], autolag = 'AIC')

print("1. ADF : ",dftest[0])
print("2. P-Value : ", dftest[1])
print("3. Num Of Lags : ", dftest[2])
print("4. Num Of Observations Used For ADF Regression and Critical Values Calculation :", dftest[3])
print("5. Critical Values :")
for key, val in dftest[4].items():
    print("\t",key, ": ", val)
```

```python
df = df.reset_index()
date = df[['date']]
X = df.drop(['Cases', 'date'], axis=1)
y = df[['Cases']]
y = y.astype('float')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression


regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
```

```python
regression_model.score(X_test, y_test)
y_predict = regression_model.predict(X_test)
```

```python
#most of the scatter is at the beginning of the pandemic
#after that predictions are very close 
plt.plot(y_test - y_predict,marker='o',linestyle='')
plt.title('actual minus predicted cases global')
```

```python
#calculate relative error for global cases
y_test = list(y_test['Cases'])
flattened = pd.DataFrame(y_predict) 
flattened = flattened[0]
sum = 0
for i in range(len(y_test)):
    sum = sum + (abs(y_test[i] - flattened[i]) / y_test[i])
sum = sum / len(y_test)
print('relative error: ', sum)
```
